```python
#sample line chart

import matplotlib.pyplot as plt

x = [1,2,3,5]
y = [10,20,25,30]

plt.plot(x, y)
plt.show()
```

```python
#sample line chart with title and x-y axis labels

plt.plot(x, y)

plt.title("Sales Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.show()
```

```python
#cutomised line chart -- trend analysis

plt.figure(figsize=(6,3)) #controls charts size
plt.plot(x, y, marker='o', linestyle='-.', color='m', linewidth=2, markersize=10)


plt.title("Sales Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.show()
```

```python
#Multiple line charts

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y1 = [10,20,30,40,50]
y2 = [20,40,60,80,100]

plt.plot(x, y1, label='Sales 2024')
plt.plot(x, y2, label='Sales 2025')

plt.title("Sales Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.legend()
plt.show()
```

```python
#simple bar chart -- comparing categories

categories = ["A","B","C"]
values = [10,20,15]

plt.bar(categories, values)
plt.title("Sales")
plt.xlabel("Categories")
plt.ylabel("Values")

plt.show()
```

```python
#simple histogram chart -- data distribution

import random
data = [random.randint(0,10) for _ in range(100)]

#print(data)
plt.hist(data, bins=10)
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

```python
#simple pie chart example --

x = [10,20,30,40]
labels = ["A","B","C","D"]

plt.pie(x, labels=labels, autopct='%1.1f%%', startangle=90,explode = [0,0.1,0,0],shadow = 'True')
plt.title("Pie Chart")

plt.show()
```

```python
#sample scatter plot example -- relationship between variables

x = [5,7,8,7,2,17,2,9,4,11,12]
y = [99,86,87,78,68,58,61,63,54,23,48]

plt.scatter(x, y, color='m')
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# **Subplot**
---
> With the subplot() function you can draw multiple plots in one figure

> plt.subplot(1, 2, 1)    
> the figure has 1 row, 2 columns, and this plot is the first plot.


```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("Sales")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("Revenue")

plt.suptitle("Gich giligili") #Super Title -- can add a title to the entire figure with the suptitle()
plt.show()
```

# **Matplotlib with Pandas**

---



```python
#Matplotlib with pandas

import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Name': ['D', 'E', 'F', 'G'],
    'Age': [25, 30, 35,50]}
df = pd.DataFrame(data)

plt.bar(df['Name'], df['Age'])
plt.title('Age Distribution')
plt.xlabel('Name')
plt.ylabel('Age')
plt.show()
```

# Saving charts
---



```python
#saving chart

import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Name': ['D', 'E', 'F', 'G'],
    'Age': [25, 30, 35,50]}
df = pd.DataFrame(data)

plt.bar(df['Name'], df['Age'])
plt.title('Age Distribution')
plt.xlabel('Name')
plt.ylabel('Age')
plt.savefig('age_distribution.png')
plt.show()
```
