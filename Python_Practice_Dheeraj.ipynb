# ***args and** ****kwargs**
---
> They are used to pass variable number of arguments to a function.
> When you don’t know how many inputs will come.


```python
#without *args usage, limited positional arguments
def total_1(a,b,c):
  return a*b+c

result = total_1(5,2,3)
print(result)


#*args usage for variable number of arguments
def total(*args):
  print(sum(args))

total(1,2,3)
total(1,2,3,4)
```

```python
#without *kwargs usage, limited keyword arguments
def print_info1(name, age):
    print(f"Name: {name}, Age: {age}")

print_info1("Dheeraj", 28)

#**kwargs usage for variable number of keyword arguments
def print_info(**kwargs):
    print(kwargs)

print_info(name="Dheeraj", age=28, city="BLR", country="India")
```

```python
#using both *args and **kwagrs together

def demo(a, *args, **kwargs):
    print("a:", a)
    print("args:", args)
    print("kwargs:", kwargs)

demo(10, 20, 30, name="Dheeraj", city="BLR")
```

# **Decorators**

---
> A decorator is a function that modifies another function’s behavior without changing its code



```python
#Without decorator

import time

def brew_filtercoffee():
  start_time = time.time()
  time.sleep(2)
  print("Brewing filter coffee")
  time.sleep(5)
  print("Filter coffee ready")
  end_time = time.time()
  print(f"Time taken: {end_time - start_time} seconds")

brew_filtercoffee()
```

```python
#With decorator

import time

#wrapper function
def dec_timer(base_fn):
  def wrapper():
    start_time = time.time()
    base_fn()
    end_time = time.time()
    print(f"Time taken: {end_time - start_time} seconds")
  return wrapper

#using @decorator syntax
@dec_timer
#base function
def brew_filtercoffee():
  time.sleep(1)
  print("Brewing filter coffee")
  time.sleep(4)
  print("Filter coffee ready")

@dec_timer
def brew_espresso():
  time.sleep(1)
  print("Brewing espresso")
  time.sleep(2)
  print("Espresso ready")

brew_filtercoffee()
brew_espresso()

#applying decorator manually
#dec_brew_filtercoffee = dec_timer(brew_filtercoffee)
#dec_brew_filtercoffee()
```

```python
#Decorator function with *args for flexible number of arguments

import time
from datetime import timedelta, datetime

#wrapper function
def dec_timer(func):
  def wrapper(*args):
    start_time = time.time()
    result = func(*args)
    end_time = time.time()
    print(f"Time taken: {end_time - start_time} seconds")
    return result
  return wrapper


@dec_timer
def brew_coffee(coffee_type,sleep_time):
  time.sleep(sleep_time)
  print("Brewing coffee")
  time.sleep(sleep_time)
  print("Coffee is ready")


@dec_timer
def brew_espresso(type, sleep_time):
  time.sleep(sleep_time)
  print("Brewing espresso")
  time.sleep(sleep_time)
  print("Espresso ready")
  return f"Drink another espresso by {datetime.now() + timedelta(minutes=60)}"

brew_coffee("filter",2)
brew_espresso("black",3)
```

## **Generator**

---
> A generator is a function that returns values one at a time using yield instead of return




```python
#Difference between RESULT and YIELD

#after return statement, function terminates the execution
def gen_values():
  return 1
  print("This will not be printed")

print(gen_values())


#yeild just pauses the exectution and returns generator objects - sequence of values
def gen_values1():
  yield 1
  yield 2
  print("This will be printed")

for i in gen_values1():
  print(i)
```

```python
#Prime number list without generator


def get_primes_list(start, end):
  primes = []
  for num in range(start, end + 1):
    if num < 2:
      continue
    is_prime = True
    for i in range(2, num):
      if num % i == 0:
        is_prime = False
        break
    if is_prime:
        primes.append(num)
  return primes

get_primes_list(1,50)
```

```python
#Prime number list with generator

def gen_primes(start, end):
  for num in range(start, end + 1):
    if num < 2:
      continue
    is_prime = True
    for i in range(2, num):
      if num % i == 0:
        is_prime = False
        break
    if is_prime:
        yield num

for value in gen_primes(30,50):
  if value % 10 == 7:
    print(value)
    break
```

```python
class Dog:
    def __init__(self, name):
        # self.name is an instance attribute
        self.name = name

    def speak(self):
        # Accessing the instance attribute using self
        print(f"{self.name} says Woof!")

# Create an instance
dog = Dog("Buddy")

# Call an instance method
dog.speak()
# Output: Buddy says Woof!
```

```python
#Types of Methods

class Student:
    school_name = "ABC School"   # Class variable

    def __init__(self, name, marks):
        self.name = name          # Instance variable
        self.marks = marks

    # 1. Instance Method
    def display(self):
        print(f"Name: {self.name}, Marks: {self.marks}")

    # 2. Class Method
    @classmethod
    def change_school(cls, new_name):
        cls.school_name = new_name

    # 3. Static Method
    @staticmethod
    def is_pass(marks):
        return marks >= 40


# Create object
s1 = Student("Dheeraj", 75)

# Instance method call
s1.display()

# Class method call
Student.change_school("XYZ School")
print(Student.school_name)

# Static method call
print(Student.is_pass(75))
```

# **OOPs Concepts**
---

1.   Abstraction
2.   Encapsulation
1.   Inheritance
2.   Polymorphism





```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary   # encapsulation

    def display(self):
        print(self.name)

    def get_salary(self):       #new method to extract encapsulation data
        return self.__salary

class Manager(Employee):        # inheritance
    def display(self):          # polymorphism
        print("Manager:", self.name)

e1 = Employee("Dheeraj", 50000)

print(e1.name)
print(e1.get_salary())
```

```python
#Inheritance example

class human:                         #parent class
  def __init__(self,name, city):
    self.name = name
    self.city = city

class child(human):                 #calling parent class
  def __init__(self,name, city, age):
    super().__init__(name, city)          #calling parent method and passing 'city'
    self.age = age                  #defining new attribute in the child method

c1 = child("Dheeraj", "Bangalore", 29) # Added 'city' argument
print(c1.name,c1.age, c1.city)


# -----------------------------------

class Animal:
    def speak(self):
        print("Animal makes sound")

class Dog(Animal):        #call parent class
    def speak(self):
        super().speak()   # call parent method
        print("Dog barks")


d = Dog()
d.speak()
```

```python
#Single Inheritance


class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):   # inherits from Animal
    def bark(self):
        print("Dog barks")

d = Dog()
d.speak()   # from parent
d.bark()    # from child
```

```python
#Multiple Inheritance

class Father:
    def skills(self):
        print("Father: Driving")

class Mother:
    def skills(self):
        print("Mother: Cooking")

class Child(Father, Mother):
    pass

c = Child()
c.skills()
```

```python
#Multi-level Inheritance


class Animal:
    def eat(self):
        print("Eating")

class Dog(Animal):
    def bark(self):
        print("Barking")

class Puppy(Dog):
    def weep(self):
        print("Weeping")

p = Puppy()
p.eat()    # from Animal
p.bark()   # from Dog
p.weep()   # from Puppy
```

> Getter → method to read/access data           
> Setter → method to modify/update data


```python
#Traditional way of using "getter" and "setter"


class Employee:
    def __init__(self, salary):
        self.__salary = salary   # private variable

    # Getter
    def get_salary(self):
        return self.__salary

    # Setter
    def set_salary(self, salary):
        if salary > 0:
            self.__salary = salary
        else:
            print("Invalid salary")


e = Employee(50000)

print(e.get_salary())   # getter

e.set_salary(7000)     # setter
print(e.get_salary())
```

```python
#Pythonic way of using "getter" and "setter"


class Employee:
    def __init__(self, salary):
        self.__salary = salary

    @property                        #getter
    def salary(self):
        return self.__salary

    @salary.setter                   #setter
    def salary(self, value):
        if value > 0:
            self.__salary = value
        else:
            print("Invalid salary")


e = Employee(50000)

print(e.salary)    # getter (no method call)

e.salary = 60000   # setter (no method call)
print(e.salary)
```
